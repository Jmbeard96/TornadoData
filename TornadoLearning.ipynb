```python
import pandas as pd

file = "Tornadoes_SPC_1950to2015.csv"
c = pd.read_csv(file)

c.head()
```

```python
c.drop('om', axis=1, inplace=True)
c.drop('yr', axis=1, inplace=True)
c.drop('mo', axis=1, inplace=True)
c.drop('dy', axis=1, inplace=True)
c.drop('date', axis=1, inplace=True)
c.drop('time', axis=1, inplace=True)
c.drop('tz', axis=1, inplace=True)
c.drop('st', axis=1, inplace=True)
c.drop('stf', axis=1, inplace=True)
c.drop('stn', axis=1, inplace=True)
c.drop('slat', axis=1, inplace=True)
c.drop('slon', axis=1, inplace=True)
c.drop('elat', axis=1, inplace=True)
c.drop('elon', axis=1, inplace=True)
c.drop('fc', axis=1, inplace=True)
```
